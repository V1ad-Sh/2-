# Одновимірна лінійна регресія
Мої показники за таблицею №3:
а = 3
b = 0
c = 1
d = 2

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

а = 3
b = 0
c = 1
d = 2

# Генерація даних:
np.random.seed(42)  # Для відтворюваності результатів
x = np.arange(0, 20)
y = a * x + b + np.random.normal(0, 1+c, 20)  # y = 3x + шум

# Перетворення x у формат, який приймає модель
x = x.reshape(-1, 1)

# Створення та навчання моделі лінійної регресії
model = LinearRegression()
model.fit(x, y)

# Виведення коефіцієнтів моделі
a = model.coef_[0]
b = model.intercept_
print(f"a = {a:.2f}, b = {b:.2f}")

# Виведення коефіцієнта детермінації R²
r2 = model.score(x, y)
print(f"R² = {r2:.4f}")

# Виконання прогнозу в довільній точці, наприклад, x = 25
x_new = np.array([[25]])
y_pred = model.predict(x_new)
print(f"Прогноз для x = {x_new[0][0]}: y = {y_pred[0]:.2f}")

# Підготовка даних для побудови лінії регресії
X_plot = np.linspace(0, 25, 100).reshape(-1, 1)
Y_plot = model.predict(X_plot)

# Побудова графіку
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Дані')
plt.plot(X_plot, Y_plot, color='red', label='Лінійна регресія')
plt.scatter(x_new, y_pred, color='green', marker='x', s=100, label='Прогноз')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Одновимірна лінійна регресія')
plt.legend()
plt.grid(True)
plt.show()
```
